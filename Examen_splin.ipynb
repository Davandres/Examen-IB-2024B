```python
xs = [-1, 0, 1]
ys = [1, 5, 3]
import matplotlib.pyplot as plt

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dataset")
plt.show()
```

```python
def Spline(x: float, x0: float, pars: dict[int, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import numpy as np

s = [
    {"a": 1, "b": 2,  "c": 6.28, "d": -4.29},
    {"a": 5, "b": 2, "c": -12.57, "d": 8.57},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    y = Spline(_x, x_i, s[i])

    plt.plot(_x, y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines")
plt.show()
```
